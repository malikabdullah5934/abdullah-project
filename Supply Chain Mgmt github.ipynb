```python
# Calculate average order quantities
avg_order_quantity = df['Order Fulfillment Rate (%)'].mean()

# Find the maximum order quantity and the associated product type
max_order_quantity = df['Order Fulfillment Rate (%)'].max()
max_order_quantity_product = df[df['Order Fulfillment Rate (%)'] == max_order_quantity]['Company Name'].values[0]

# Find the minimum order quantity and the associated product type
min_order_quantity = df['Order Fulfillment Rate (%)'].min()
min_order_quantity_product = df[df['Order Fulfillment Rate (%)'] == min_order_quantity]['Company Name'].values[0]

# Print results
print(f"Average Order Quantity: {avg_order_quantity:.2f}")
print(f"Maximum Order Quantity: {max_order_quantity} (Company Name: {max_order_quantity_product})")
print(f"Minimum Order Quantity: {min_order_quantity} (Company Name: {min_order_quantity_product})")


```

```python
# Calculate average price
mean_unit_qty = df['Cost of Goods Sold (COGS)'].mean()

# Calculate standard deviation
std_dev = df['Transportation Cost Efficiency (%)'].std()

# Find maximum and minimum prices with dates
max_price = df['Transportation Cost Efficiency (%)'].max()
max_price_rows = df[df['Transportation Cost Efficiency (%)'] == max_price]
max_price_date = max_price_rows['Lead Time (days)'].values[0] if not max_price_rows.empty else None

min_price = df['Transportation Cost Efficiency (%)'].min()
min_price_rows = df[df['Transportation Cost Efficiency (%)'] == min_price]
min_price_date = min_price_rows['Lead Time (days)'].values[0] if not min_price_rows.empty else None

# Calculate average daily volume
avg_volume = df['Supplier Count'].mean()

# Print results
print(f"Average Price: ${mean_unit_qty:.2f}")
print(f"Price Volatility (Standard Deviation): ${std_dev:.2f}")
print(f"Maximum Price: ${max_price:.2f} (Date: {max_price_date})")
print(f"Minimum Price: ${min_price:.2f} (Date: {min_price_date})")
print(f"Average Daily Volume: {avg_volume:.2f} units")
```

```python
# Drop the 'Revenue Growth Rate out of (15)' column
X = df.drop('Revenue Growth Rate out of (15)', axis=1)
```

```python
y = df['Customer Satisfaction (%)']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
# Check the data types of X_train
print(X_train.info())

# Step 1: Handle date columns (if applicable)
# Identify and convert date columns
for col in X_train.select_dtypes(include=['object']):
    try:
        X_train[col] = pd.to_datetime(X_train[col])
        X_train[col] = X_train[col].astype(int) / 10**9  # Convert to Unix timestamp
    except Exception as e:
        print(f"Column '{col}' is not a date column.")

# Step 2: Handle categorical columns
# Convert non-numeric and non-date columns to dummy variables
categorical_columns = ['Product type', 'SKU', 'Availability', 'Customer demographics', 
                       'Shipping carriers', 'Supplier name', 'Location', 'Transportation modes', 'Routes']
X_train = X_train.drop(columns=categorical_columns, errors='ignore')  # Remove any unwanted non-numeric columns
X_train = pd.get_dummies(X_train, drop_first=True)  # Convert categorical columns to dummy variables

# Step 3: Ensure the data is numeric
print(X_train.info())

# Step 4: Fit the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print("Model training successful!")
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer object that replaces NaN with the column mean
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform both the training and testing data
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Train the model using the imputed data
model = LinearRegression()
model.fit(X_train_imputed, y_train)

# Predicting values using the trained model on X_test
y_predicted = model.predict(X_test_imputed)

# Calculate residuals (difference between actual and predicted values)
test_residual = y_test - y_predicted

# Plot residuals to check for any patterns
import seaborn as sns
sns.displot(test_residual, bins=25, kde=True)


```
